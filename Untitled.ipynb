```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from pprint import pprint
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
from functions import tech, human, natural, creative, file_names, linii_without_end, linii_without_start
```

```python
data_universities_lst = []
for filename, end, start in zip(file_names, linii_without_end, linii_without_start):
    #print(filename)
    html_code = open(filename, 'r').read()
    lst = html_code.split(end)
    data_universities_without_end = lst[0]
    lst_1 = data_universities_without_end.split(start)
    data_universities = lst_1[1]
    data_universities_lines = data_universities.split('\n')
    data_universities_lines = [line for line in data_universities_lines if '    ' in line]
    data_universities_lines = [i.split('    ')[1] for i in data_universities_lines]
    #pprint(data_universities_lines[:20])
    #pprint(data_universities_lines[-21:])
    
    len_data_universities_lines = len(data_universities_lines)
    
    if filename in ['2011.txt']:
        for i in range(0, len_data_universities_lines, 11):
            oblast = data_universities_lines[i]

            university = data_universities_lines[i+1]

            ball = data_universities_lines[i+3]
            
            budget_students = data_universities_lines[i+5]
            data_universities_lst.append([oblast, university, ball, filename[:4], budget_students])
    
    if filename in ['2012.txt']:
        for i in range(0, len_data_universities_lines, 11):
            oblast = data_universities_lines[i]

            university = data_universities_lines[i+1]

            ball = data_universities_lines[i+2]
            
            budget_students = data_universities_lines[i+4]
            data_universities_lst.append([oblast, university, ball, filename[:4], budget_students])
        
    if filename in ['2013.txt', '2014.txt']:
        for i in range(0, len_data_universities_lines, 10):
            oblast = data_universities_lines[i]

            university = data_universities_lines[i+1]

            ball = data_universities_lines[i+2]
            
            budget_students = data_universities_lines[i+4]
            data_universities_lst.append([oblast, university, ball, filename[:4], budget_students])

    if filename in ['2015.txt']:
        for i in range(0, len_data_universities_lines, 11):
            oblast = data_universities_lines[i]
            
            university = data_universities_lines[i+1]
            
            ball = data_universities_lines[i+2]
            
            budget_students = data_universities_lines[i+3]
            data_universities_lst.append([oblast, university, ball, filename[:4], budget_students]) 
    if filename in ['2016.txt']:
        for i in range(0, len_data_universities_lines, 9):
            oblast = data_universities_lines[i]

            university = data_universities_lines[i+1]

            ball = data_universities_lines[i+2]
            
            budget_students = data_universities_lines[i+4]
            data_universities_lst.append([oblast, university, ball, filename[:4], budget_students])
    if filename in ['2017.txt', '2018.txt', '2019.txt']:
        for i in range(0, len_data_universities_lines, 7):
            oblast = data_universities_lines[i]

            university = data_universities_lines[i+1]

            ball = data_universities_lines[i+2]
            
            budget_students = data_universities_lines[i+4]
            data_universities_lst.append([oblast, university, ball, filename[:4], budget_students])
        
    #pprint(data_universities_lst)
```

```python
df = pd.DataFrame(data_universities_lst)
df.columns = ['oblast', 'university', 'ball', 'year', 'budget_student']
df = df.sort_values(by=['university', 'oblast', 'year'])
df['ball'] = [float(i) if i != '' else '' for i in df['ball']]
df['budget_student'] = [float(i) if i != '' else '' for i in df['budget_student']]
df = df[df['ball'] != '']
df.head()
fields = []
for i in df['oblast']:
    if i in human:     
        fields.append('human') 
    elif i in tech:     
        fields.append('tech') 
    elif i in natural:     
        fields.append('natural') 
    elif i in creative:     
        fields.append('creative') 
    else:
        fields.append('trash')
df['field'] = fields
df.head()
```

```python
large_university = []
for university, group in df.groupby('university'):
    if group['budget_student'].sum() >=5000:
        large_university.append(university)
#pprint(large_university)
df_large = df[df['university'].isin(large_university)]
df_small = df[~df['university'].isin(large_university)]
print(len(set(df['university'])), len(set(df_large['university'])), len(set(df_small['university'])))
```

```python
good_university = []
for university, group in df.groupby('university'):
    if group['ball'].mean() >=75:
        good_university.append(university)
#pprint(large_university)
df_good = df[df['university'].isin(good_university)]
df_bad = df[~df['university'].isin(good_university)]
print(len(set(df['university'])), len(set(df_good['university'])), len(set(df_bad['university'])))
```

```python
quality = []
for i in df['university']:
    if i in good_university:     
        quality.append('good') 
    else:     
        quality.append('bad') 
df['quality'] = quality
df.head()
```
