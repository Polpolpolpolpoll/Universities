```python
import matplotlib.pyplot as plt
import pandas as pd
from pprint import pprint
file_names = ['2011.txt', '2012.txt', '2013.txt', '2014.txt', '2015.txt', '2016.txt', '2017.txt', '2018.txt', '2019.txt']
linii_without_end = ['\n<span id="line198572"', '\n<span id="line208034"', '\n<span id="line178017"', '\n<span id="line193702"', '\n<span id="line168945"', '\n<span id="line134083"', '\n<span id="line104273"', '\n<span id="line131510"', '\n<span id="line129229"']
linii_without_start = ['<span id="line1353"></span>',
                       '<span id="line1325"></span>',
                       '<span id="line1316"></span>',
                       '<span id="line1321"></span>',
                       '<span id="line1333"></span>',
                       '<span id="line1324"></span>',
                       '<span id="line1305"></span>',
                       '<span id="line1333"></span>',
                       '<span id="line1329"></span>']
data_universities_lst = []
for filename, end, start in zip(file_names, linii_without_end, linii_without_start):
    print(filename)
    html_code = open(filename, 'r').read()
    lst = html_code.split(end)
    data_universities_without_end = lst[0]
    lst_1 = data_universities_without_end.split(start)
    data_universities = lst_1[1]
    data_universities_lines = data_universities.split('\n')
    data_universities_lines = [line for line in data_universities_lines if '    ' in line]
    data_universities_lines = [i.split('    ')[1] for i in data_universities_lines]
    #pprint(data_universities_lines[:20])
    #pprint(data_universities_lines[-21:])
    
    len_data_universities_lines = len(data_universities_lines)
    
    if filename in ['2011.txt']:
        for i in range(0, len_data_universities_lines, 11):
            oblast = data_universities_lines[i]

            university = data_universities_lines[i+1]

            ball = data_universities_lines[i+3]
            data_universities_lst.append([oblast, university, ball, filename[:4]])
    
    if filename in ['2012.txt']:
        for i in range(0, len_data_universities_lines, 11):
            oblast = data_universities_lines[i]

            university = data_universities_lines[i+1]

            ball = data_universities_lines[i+2]
            data_universities_lst.append([oblast, university, ball, filename[:4]])
        
    if filename in ['2013.txt', '2014.txt']:
        for i in range(0, len_data_universities_lines, 10):
            oblast = data_universities_lines[i]

            university = data_universities_lines[i+1]

            ball = data_universities_lines[i+2]
            data_universities_lst.append([oblast, university, ball, filename[:4]])

    if filename in ['2015.txt']:
        for i in range(0, len_data_universities_lines, 11):
            oblast = data_universities_lines[i]
            
            university = data_universities_lines[i+1]
            
            ball = data_universities_lines[i+2]
            data_universities_lst.append([oblast, university, ball, filename[:4]]) 
    if filename in ['2016.txt']:
        for i in range(0, len_data_universities_lines, 9):
            oblast = data_universities_lines[i]

            university = data_universities_lines[i+1]

            ball = data_universities_lines[i+2]
            data_universities_lst.append([oblast, university, ball, filename[:4]])
    if filename in ['2017.txt', '2018.txt', '2019.txt']:
        pprint(data_universities_lines[:20])
        for i in range(0, len_data_universities_lines, 7):
            oblast = data_universities_lines[i]

            university = data_universities_lines[i+1]

            ball = data_universities_lines[i+2]
            data_universities_lst.append([oblast, university, ball, filename[:4]])
        
    #pprint(data_universities_lst)
```

```python
df = pd.DataFrame(data_universities_lst)
df.columns = ['oblast', 'university', 'ball', 'year']
df = df.sort_values(by=['university', 'oblast', 'year'])
df['ball'] = [float(i) if i != '' else '' for i in df['ball']]
df = df[df['ball'] != '']
df.head()
```

```python
tech = ['Авиационная и ракетно-космическая техника',
 'Авиационная и ракетно-космическая техника (теория)',
 'Авиационные системы (эксплуатация)',
 'Автоматика и управление',
 'Информатика и вычислительная техника',
 'Информационная безопасность',
 'Математика',
 'Машиностроение',
 'Приборостроение и оптотехника',
 'Технологии легкой промышленности',
 'Технологические машины и оборудование',
 'Физика',
 'Электронная техника, радиотехника и связь',
 'Энергетика и энергетическое машиностроение',
 'Ядерные физика и технологии']

human = ['Экономика',  
'Реклама и связи с общественностью',
'Юриспруденция',
'Искусства и гуманитарные науки',
'Культурология',
'Религиеведение и теология',
'Журналистика и литературное творчество',
'Теория искусств',
'Социология',
'Сестринское дело',
'Сестринское дело2',
'Педагогическое образование',
'Актерское исполнительство',
'Лингвистика и иностранные языки',
'Социальная работа',
'Государственное и муниципальное управление',
'История',
'Библиотеки и архивы',
'Востоковедение и африканистика',
'Менеджмент',
'Международные отношения',
'Политология',
'Психология',
'Психолого-педагогическое и специальное (дефектологическое) образование',]

df_tech = df[df['oblast'].isin(tech)]
medium_bally_tech = []
for year, group in df_tech.groupby('year'):
    medium_bally_tech.append(group['ball'].mean())
    
df_human = df[df['oblast'].isin(human)]
medium_bally_human = []
for year, group in df_human.groupby('year'):
    medium_bally_human.append(group['ball'].mean())

x = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]

plt.plot(x, medium_bally_tech, label='tech')
plt.plot(x, medium_bally_human, label='human')
plt.legend()
```
